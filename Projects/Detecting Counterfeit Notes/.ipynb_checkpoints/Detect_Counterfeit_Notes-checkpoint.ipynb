## 3. Using a Classification Algorithm

### 3.2 Performing K-means clustering

Since we need to differentiate between genuine and fake notes, we naturally select n_clusters = 2 in our algorithm.

```python
# Import the kmeans algorithm
from sklearn.cluster import KMeans

# Create a k-means clustering model with 2 clusters
kmeans = KMeans(init='random', n_clusters=2, n_init=10)

#Removing the is_genuine column from our X_scaled variable

# Fit the data to the model
kmeans.fit(X_scaled)

# Determine which clusters each data point belongs to:
clusters =  kmeans.predict(X_scaled)

# Add cluster number to the original data
X_scaled_clustered = pd.DataFrame(X_scaled, columns=df2.columns, index=df2.index)
X_scaled_clustered['cluster'] = clusters

X_scaled_clustered.head()
```

### 3.4 Scoring our classification algorithm

```python
#First we inject our clusters into our main dataframe
df_clustered = df.copy()
df_clustered["cluster"] = clusters

#Number of correctly predicted notes:
len(df_clustered[df_clustered["is_genuine"] == df_clustered["cluster"]])

#Calcuting the accuracy of our model
acc = len(df_clustered[df_clustered["is_genuine"] == df_clustered["cluster"]]) / len(df_clustered)

#Calculating our true/false positive/negative values
tp = len(df_clustered.loc[(df_clustered["is_genuine"] == 1)&(df_clustered["cluster"]==1)])
tn = len(df_clustered.loc[(df_clustered["is_genuine"] == 0)&(df_clustered["cluster"]==0)])
fp = len(df_clustered.loc[(df_clustered["is_genuine"] == 0)&(df_clustered["cluster"]==1)])
fn = len(df_clustered.loc[(df_clustered["is_genuine"] == 1)&(df_clustered["cluster"]==0)])

#Calculating scoring metrics
prec = tp/(tp + fp)
recall = tp/(tp+fn)
specificity = tn/(fp+tn)
f1_score = tp/(tp + 0.5*(fp + fn))

#Displaying the results
kmeans_results = pd.DataFrame({"algorithm": "K-means clustering","Accuracy": acc, "Precision": prec,
                               "Recall": recall, "Specificity": specificity, "F1_Score": f1_score}, index=range(1))

kmeans_results
```

```python
#Creating a confusion matrix
confusion_matrix = np.array([[tn,fp],[fn,tp]], dtype="int64")

#Displaying our matrix
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

## 4. Building a Logistic Regression Model

### 4.1 First basic model

We will generate a model based on the 2 variables that are most strongly correlated with our target is_genuine variable : margin_low and margin_up

```python
#Importing logistic regression model
import statsmodels.formula.api as smf

#First basic model
#We select variables that are the most correlated with is_genuine
model = smf.logit('is_genuine ~ margin_low + margin_up', data = df)
results = model.fit()
results.summary()
```

Our Pseudo R Squared is very good at 0.944, and we see that our coefficients p values are under < 0.05 so we can reject the null hypothesis that these coefficients do not predict our target variable.

### 4.2 Probabilities histogram

```python
#Calculating the probabilities
proba = 1 / (1 + np.exp( - results.fittedvalues ))

#Plotting them with a histogram
sns.histplot(proba)
plt.show()
#Our model has a good histogram of probabilities, most values are at 0 or 1
```

Our histplot seems to have a good repartition of probabilities are the immense majority of values are either close to 0 or 1.

### 4.3 Determining the best model using ROC AUC

We have a good first model draft but we want to verify that it is the best by testing several formulas.

Note : In a real life analysis, this formula list would have been drastically diminished because it would require a lot of machine time to run the models with an important dataset.

We will use **2 criteria : Accuracy and ROC AUC Score** to figure out the effectiveness of our models. Accuracy is simply the sum of correctly predicted observations divided by the total number of observations while the ROC AUC score also takes into account the slight imbalance between the number of genuine (100) and not genuine (70) notes in our dataset.

We have to use the **Broyden-Fletcher-Goldfarb-Shanno algorithm (bfgs)** here instead of the more classical Newton algorithm because we have a model that fits our dataset perfectly. The Newton method is not able to handle such a model because it tries to invert a Singular Matrix, which is not possible.

```python

#Defining the accuracy
def accuracy(result, df):
    #We define the accuracy thanks to the pred_table function of sklearn
    return (result.pred_table()[0,0]+result.pred_table()[1,1])/len(df)

#Importing ROC AUC SCORE
from sklearn.metrics import roc_auc_score

#Generating a list of all test formulas
formulas = [
    'is_genuine ~ length',
    'is_genuine ~ length + margin_up',
    'is_genuine ~ length + margin_low',
    'is_genuine ~ length + margin_low + margin_up',
    'is_genuine ~ length + margin_low + margin_up + height_right',
    'is_genuine ~ margin_low + margin_up',
    'is_genuine ~ margin_low + margin_up + diagonal',
    'is_genuine ~ margin_low + margin_up + height_left + height_right',
    'is_genuine ~ margin_low + margin_up + height_left + height_right + diagonal',
    'is_genuine ~ height_left + height_right',
    'is_genuine ~ length + diagonal',
    'is_genuine ~ diagonal + height_left + height_right'
]

#Creating an empty array to store results
scores = []

#Iterating over all forumlas
for formula in formulas:
    #Defining the model, we take a maximum number of iteration = 60 to allow the model to converge
    results = smf.logit(formula, df).fit(method = 'bfgs', maxiter = 60)
    yhat = results.predict()
    #Adding the scores to our array
    scores.append( { 'model': formula,
        'accuracy': accuracy(results, df),
        'ROC AUC Score': roc_auc_score(df["is_genuine"],yhat)}
    )

#Turning the array into a dataframe for easier visualization
scores = pd.DataFrame(scores)

print(scores)
```

**2 models seem to have a perfect accuracy and ROC_AUC_SCORE** :

- is_genuine ~ length + margin_low + margin_up 
- is_genuine ~ length + margin_low + margin_up + height_right

The first reaction would be to delete height_right from our formula since it does not add anything to the model, but we will first validate our findings with cross validation.

### 4.4 Cross validating the formulas

Our models need not only work with our current dataset but more importantly they need to be able to **predict the genuity of notes**. We will use cross_validation to find the best model.

```python
#Initializing random variable
np.random.seed(8)

#Sampling the dataset, we take a default 70/30 split here
train_index = df.sample(frac = 0.7).index
# Create the train and test subset
train = df.loc[df.index.isin(train_index)]
test = df.loc[~df.index.isin(train_index)]

#Initializing our scores array that will contain all result variables
scores = []

for formula in formulas:
    #as mentionned previously, we use the bfgs method instead of Newton
    #to avoid Singular Matrix Error
    results = smf.logit(formula, train).fit(method = 'bfgs', maxiter=50)
    #Calculating our probabilities
    yhat = results.predict(test)
    #Adding the scores to our array
    scores.append( { 'model': formula,
        #Here we calculate the "ROC AUC error" by substracting 1 by our ROC AUC scores
        #This allows us to find the best model more easily
        'ROC_error_train': 1-roc_auc_score(train.is_genuine, results.predict(train)),
        'ROC_error_test': 1-roc_auc_score(test.is_genuine,yhat)}
    )
#Converting our scores array into a Dataframe    
scores = pd.DataFrame(scores)

#Selecting average ROC AUC Score and accuracy by model
scores = scores.groupby("model").mean().reset_index()

#Creating a total error column to order our results
scores["total_average_error"] = scores.ROC_error_test + scores.ROC_error_train

#Displaying our results starting with the one with the smallest average error
print(scores.sort_values(by="total_average_error"))
```

```python
#Let's quickly build a scatter plot of scores for the train and test subsets
plt.scatter(scores.index, scores.ROC_error_train, label = 'train'  )
plt.scatter(scores.index, scores.ROC_error_test, label = 'test'  )
sns.despine()
plt.legend()
plt.show()
```

We have chosen here to display ROC error (the difference between ROC and 1) in order to more easily display our results on a graph.

The visualization does not provide a lot of information since a lot of these models have ROC errors very close to 0.

It is clear with the scores dataframe though that our "is_genuine ~ length + margin_low + margin_up" and "is_genuine ~ length + margin_low + margin_up + height_right" models have a mean error of 0, which means that they perfectly predict our dataset.

If we had to choose right now, we would **select the first model because the added variable adds complexity but does not improve the results**.

### 4.5 K-fold (4) cross validation

We will now try **4-fold cross validation** to verify one last time that we have the good model, using the BFGS method once again. This K fold validation will prevent a potential bias stemming from the initialization variable of the np.random function.

```python
#Initializing random variable
np.random.seed(8)
#Initializing K
K = 4
#Sampling the Dataset
df = df.sample(frac=1)
indexes = np.array_split(list(df.index),K)

#Initializing our scores array that will contain all result variables
scores = []

for formula in formulas:
    for i in range(K):
        #Defining our test and train subsets
        test_index  = indexes[i]
        train_index = [idx for idx in list(df.index) if idx not in test_index]
        train = df.loc[df.index.isin(train_index)]
        test  = df.loc[~df.index.isin(train_index)]
        #as mentionned previously, we use the bfgs method instead of Newton
        #to avoid Singular Matrix Error
        results = smf.logit(formula, train).fit(method = 'bfgs', maxiter = 50)
        #Calculating our probabilities
        yhat = results.predict(test)
        #Adding the scores to our array
        scores.append( { 'model': formula,
        #Here we calculate the "ROC AUC error" by substracting 1 by our ROC AUC scores
        #This allows us to find the best model more easily
            'ROC_error_train': 1-roc_auc_score(train.is_genuine, results.predict(train)),
            'ROC_error_test': 1-roc_auc_score(test.is_genuine,yhat)}
    )
#Converting our scores array into a Dataframe    
scores = pd.DataFrame(scores)

#Selecting average ROC AUC Score and accuracy by model
scores = scores.groupby("model").mean().reset_index()

#Creating a total error column to order our results
scores["total_error"] = scores.ROC_error_test + scores.ROC_error_train

#Displaying our results starting with the one with the smallest average error
print(scores.sort_values(by="total_error"))
```

**Our "is_genuine ~ length + margin_low + margin_up" has a total error of 0, we can now safely conclude that it is the best model.**

The added variable (height_right) does not add anything to the test and we will thus remove it from the equation.

### 4.6 Displaying a classification report for our model

```python
#Calculating a classification report for our best model
m = 'is_genuine ~ margin_up + margin_low + length'

from sklearn.metrics import classification_report
result = smf.logit(m, data = df).fit()
y = result.predict(df).to_frame()
y.set_axis(["prob"],axis=1, inplace=True)
y.loc[y.prob>0.5] = 1
y.loc[~(y.prob>0.5)] = 0
print(classification_report(df['is_genuine'], y))

#Seems that our model perfectly classifies our dataset
```

Once again, we conclude that our model perfectly predicts our dataset. 

Ideally, we would like to **select our model using a bigger dataset** to be 100% sure that we have selected the right model since it is not realistic to find a model with a 100% accuracy rate.

## 5. Creating a classification program

We will now create a classification algorithm that will be able to classify other notes thanks to our model. As required, this function will take as an argument the location of a csv file that contains all our data except the is_genuine variable as well as the ID of the note.

```python
def note_classification(file_path):
    #File needs to be in root folder of Jupyter notebook or full path needs to be given
    data = pd.read_csv(file_path)
    
    #Generating our model from our full dataset
    result = smf.logit(m, data=df).fit()
    
    #Setting the id column as an index for our initial file
    data.set_index("id", inplace=True)
    
    #Calculating the probability of classification for our model and creating a dataframe
    y = result.predict(data).to_frame()
    
    #Renaming the generated probability column
    y.columns = ["probability_genuine"]
    #Duplicating the column to create an is_genuine column
    y["is_genuine"] = y["probability_genuine"]
    #Setting probabilities >= 0.5 to genuine and < 0.5 to not genuine 
    y.loc[y.is_genuine >= 0.5, "is_genuine"] = True 
    y.loc[y.is_genuine < 0.5, "is_genuine"] = False
    
    #Merging the genuity with our dataset 
    data = data.merge(y, how="left", on="id")
    
    #Creating a new path to prevent overriding the source file
    new_path = file_path.replace(".csv","") + "_classified.csv"
    #Saving our classified dataframe
    data.to_csv(new_path)
    
    #Returning the classified dataframe
    return data
    
```
