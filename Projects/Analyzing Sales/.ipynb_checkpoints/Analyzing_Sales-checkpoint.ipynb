## 4. Data analysis

```python
#Defining a function to define correlation
def correlation(r):
    if abs(r) > 1: return 'Correlation wrongly calculated, cannot be > 1'
    elif abs(r) > 0.7: return 'Strong correlation'  
    elif abs(r) > 0.5: return 'Moderate Correlation' 
    elif abs(r) > 0.3: return 'Weak correlation' 
    else: return 'No significant correlation' 

#Importing statistical functions
import scipy.stats as st

#Defining eta_squared
def eta_squared(x,y):
    mean_y = y.mean()
    categories = []
    for category in x.unique():
        yi_category = y[x==category]
        categories.append({'ni': len(yi_category),
                        'mean_category': yi_category.mean()})
    TSS = sum([(yj-mean_y)**2 for yj in y])
    ESS = sum([c['ni']*(c['mean_category']-mean_y)**2 for c in categories])
    return ESS/TSS
```

#### 4.2.2 Concentration of price (Lorenz Curve)

```python
#Lorenz Curve
prices = data["price"].values
n = len(prices)
lorenz = np.cumsum(np.sort(prices)) / prices.sum()
lorenz = np.append([0],lorenz) # The Lorenz Curve begins at 0

plt.axis('equal')
xaxis = np.linspace(0-1/n,1+1/n,n+1) # There is 1 segment (of size n) for each individual, plus 1 segment at y=0. The first segment starts at 0-1/n and the last one finishes at 1+1/n
plt.plot(xaxis,lorenz,drawstyle='steps-post', label = 'Lorenz Curve')
plt.plot(xaxis,xaxis, label = 'Perfect Equality')
plt.title("Lorenz curve showing profits and number of transactions")
plt.xlabel("Cumulative share of transactions")
plt.ylabel("Cumulative share of profits")
plt.legend()
plt.show()

AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # area under the Lorenz Curve. The first segment (lorenz[0]) is halfly below O, so we divide it by 2. We do the same for the mast segment lorenz[-1]
S = 0.5 - AUC # area between 1st bisector and the Lorenz Curve
gini = 2*S # Calculation of the gini coefficient
print("Gini =",gini)
```

### 4.3Time series analysis

```python
#Calculating ordinal date from data to be able to calculate regressions and correlation coefficients
data["date_ord"] = data.date.apply(lambda x: x.toordinal())

#Turning birth info into age
data["age"] = data.date.dt.year - data["birth"]

data[data.age.isna() == True]
#Returns empty table

data.drop(columns="birth", inplace = True)

data.head()
```

```python
#Creating a subset of our data by aggregating over month in the yyyy-mm format
subset = data[["date","price"]].groupby(data.date.dt.strftime('%Y-%m')).agg(price = ('price','sum'),
                                                                           sales = ('price','count')).reset_index()

#Calculating ordinal date which will be used for regression
subset["date_ord"] = pd.to_datetime(subset['date'], format='%Y-%m', errors='coerce').apply(lambda x: x.toordinal())

subset.head()
```

```python
#Plotting a date and price time series
subset.plot(x="date", y="price",kind="line",label = "Data")
plt.title('Evolution of monthly profits')
plt.ylabel('Monthly profits (USD)')
plt.xlabel('Year-Month')
plt.show()

#We remark a sharp drop in october, but apart from that the curve seems linear
r = round(st.pearsonr(subset["date_ord"],subset["price"])[0],2)
print("r coefficient of profits and date :", r, correlation(r))
#The correlation is low because of the drop in september

#If we remove the sharp drop in october for the LR : 
subset_lr = subset[subset.date != "2021-10"]

#Creating linear regression from subset without outlier value
from sklearn.linear_model import LinearRegression

X = subset_lr["date_ord"].values.reshape(-1,1)
Y = subset_lr["price"]
lr = LinearRegression()
lr.fit(X, Y)

#Adding LR to graph
lr_val = lr.predict(subset["date_ord"].values.reshape(-1,1))

subset.plot(x="date", y="price",kind="line", label = "Data")
plt.plot(subset["date"],lr_val,color="red",linewidth=2,label="Linear Regression (October excluded)")
plt.title('Monthly profits trend')
plt.ylabel('Monthly profits (USD)')
plt.xlabel('Date')
plt.legend()
plt.show()
#Calculating r coefficient when october is excluded
r = round(st.pearsonr(subset_lr["date_ord"],subset_lr["price"])[0],2)
print("r coefficient of profits and date (October excluded) :", r, correlation(r))

#Calculating yearly growth
yearly_growth = round(subset[subset.date == "2022-02"]["price"].tolist()[0]-subset[subset.date == "2021-03"]["price"].tolist()[0],2)
print("Yearly Growth :",yearly_growth)
print("Yearly Growth Pct:", round(100*yearly_growth/subset[subset.date == "2021-03"]["price"].tolist()[0],2))
```

### 4.12 Inactive clients

```python
#Measuring inactive clients

#We will define inactive clients as clients that have not bought within 2 months
#First we will calculate the inactive date for each client

from dateutil.relativedelta import relativedelta

#We have to create a function to add months
# Defining the function
def add_months(start_date, delta_period):
  end_date = start_date + relativedelta(months=delta_period)
  return end_date
#Creating inactive_date column
client_data["inactive_date"] = client_data.max_date.apply(lambda x: add_months(x,2)).dt.date

client_data.head()
#To effectively count inactive clients, we have to remove rows with inactive dates > latest date of information
inactive_clients = client_data[client_data.inactive_date < datetime.date(2022,2,28)][["inactive_date"]].groupby("inactive_date").agg(number = ('inactive_date' , 'count')).reset_index()

#Creating a polyfit to smooth the curve
import matplotlib.dates as mdates
x = mdates.date2num(inactive_clients["inactive_date"])
y = inactive_clients["number"]
z4 = np.polyfit(x, y, 3)
p4 = np.poly1d(z4)

#Removing negative results from polyfit
pos_dates = inactive_clients[inactive_clients.inactive_date > datetime.date(2021,6,1)]["inactive_date"]
#Creating array of numdates for pos dates
xx = mdates.date2num(pos_dates)

plt.plot(pos_dates, p4(xx),color="red",linewidth=2,label="Cubic Regression")
sns.lineplot(x="inactive_date",y="number",data=inactive_clients, label="Data")
plt.title("Number of daily inactive clients")
plt.xlabel("Date")
plt.ylabel("Daily inactive clients")
plt.legend()
plt.show()
```
