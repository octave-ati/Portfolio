## 2. Preparing the data for analysis
### 2.1 Calculating Yearly Population Growth

```python
#Calculating yearly population growth

pop = pd.read_csv("Data/pop.csv")

#Changing the unit to 1 person
pop["Value"] = pop.Value*1000

#Renaming columns
pop.columns = ["x1","x2","country_code","country","x3","x4","x5","x6","year","x7","x8","pop","x9","x10"]

#Dropping useless columns
pop.drop(columns=["x1","x2","country","x3","x4","x5","x6","x7","x8","x9","x10"], inplace = True)

#Removing duplicate Chinese values (iaw project 4)
pop = pop.loc[pop["country_code"]!= 351]

#Verifying that the total population is coherent
pop_2019 = pop.loc[pop.year == 2019]["pop"].sum()

#Checking that we have no null values
pop.info()
```

```python
#Calculating yearly population growth over the 2 last years of data (2017-2019)
pop_info = pop.pivot_table(index="country_code",columns="year",values="pop").reset_index()

pop_info.columns = ["country_code","pop_2017","pop_2018","pop_2019"]

#Calculating population growth rate in percent
pop_info["pop_growth"] = 100*(pop_info.pop_2019 - pop_info.pop_2017)/(pop_info.pop_2017*2)

pop_info.head()
```

## 3. Data Analysis
### 3.1 Hierarchical clustering
#### 3.1.1 Selecting useful data for clustering

```python
#Selecting only some rows for hierarchical clustering
X = df[["country", "ps_tot","fs_tot","ps_anim_pct","pop_growth",
        "gdp_pcapita","gdp_growth","poultry_supply","chicken_imports"]].copy()

#Setting up country as an index
X.set_index("country", inplace=True)

X.head()
```

#### 3.1.2 Normalizing the data

```python
# Import the hierarchical clustering algorithm
from sklearn.cluster import AgglomerativeClustering

# Import the sklearn function
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### 3.1.3 Separating our data in 5 clusters

```python
# Create a hierarchical clustering model
hiercluster = AgglomerativeClustering(affinity='euclidean', linkage='ward', compute_full_tree=True) 

# Fit the data to the model and determine which clusters each data point belongs to:
hiercluster.set_params(n_clusters=5)
#Computing distance for dendogram
hiercluster.set_params(compute_distances = True) 
clusters = hiercluster.fit_predict(X_scaled) 
np.bincount(clusters) # count of data points in each cluster
```

#### 3.1.4 Dendogram for all studied countries

```python
from scipy.cluster.hierarchy import linkage, fcluster 
#Importing visualizing functions
from functions import *

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
plt.figure(figsize=(20,40))
plt.title('Hierarchical Clustering Dendrogram for all countries')
plt.xlabel('Distance')

#We iteratively find the value of color_threshold that plots our 5 clusters with different colors
plot_dendrogram(hiercluster,orientation="left", labels=X.index, color_threshold=11)

#Saving our figure to our working directory
plt.savefig('dendrogram.png', dpi=1200)
```

#### 3.1.5 Dendogram for each clusters

```python
# Add cluster number to the scaled data
X_scaled_clustered = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
X_scaled_clustered['cluster'] = clusters
```

```python

# Show a dendrogram, for all clusters separately

for i in range(5):
    sample = X_scaled_clustered[X_scaled_clustered.cluster==i]
    if len(sample) <= 1:
        continue
    Z = linkage(sample, 'ward') 
    names = sample.index 
    plt.figure(figsize=(10,20))
    plt.title('Hierarchical Clustering Dendrogram for Cluster {}'.format(i))
    plt.xlabel('Distance')
    dendrogram(
        Z,
        labels = names,
        orientation = "left",
    )
```

### 3.3 Studying the distribution of variables

#### 3.3.1 Searching for potential normally distributed variables

```python
#Let's generate a distribution plot of each numeric df variables to figure out which variables are normally distributed

col = df.columns.values

col = np.delete(col, np.where(col == "country"))

for i in col:
    sns.distplot(df[i])
    plt.show()
    
```

```python
#gdp growth seems to be the variable that is the most near normal distribution

sns.distplot(df.gdp_growth)
plt.xlabel("Annual GDP Growth (%) - 2019")
plt.title("Distribution of GDP Growth")
plt.show()
```

####  3.3.2 Testing normal distribution with the Kolmogorov Smirnoff Test

For this test, the null hypothesis is that the variable is normally distributed

```python
#Normalizing our variable
df['gdp_growth_norm'] = (df.gdp_growth - np.mean(df.gdp_growth)) / np.std(df.gdp_growth)

df[["gdp_growth_norm"]].describe()
```

```python
from scipy import stats
stats.kstest(df.gdp_growth_norm , 'norm') 
#The p value is under 0.05, we can reject the null hypothesis which is that the variable is normally distributed
```

```python
#We look at another variable, ps_tot, which also looks normally distributed on our pairplot
sns.distplot(df.ps_tot)
plt.xlabel("Protein Supply (g/capita/day) - 2019")
plt.title("Distribution of Protein Supply")
plt.show()
```

```python
#Normalizing the variable
df['ps_tot_norm'] = (df.ps_tot - np.mean(df.ps_tot)) / np.std(df.ps_tot)

df[["ps_tot_norm"]].describe()
```

```python
stats.kstest(df.ps_tot_norm , 'norm')

#The p value is far higher than 0.05, we cannot reject the null hypothesis and we conclude that the variable is normally distributed.
```

```python
#Testing the normal distribution of the pop growth variable
sns.distplot(df["pop_growth"])
plt.xlabel("Population growth - 2019")
plt.title("Distribution of Population Growth in all countries")
plt.show()

#Normalizing the variable
df['pop_growth_norm'] = (df.pop_growth - np.mean(df.pop_growth)) / np.std(df.pop_growth)

df[["pop_growth_norm"]].describe()
```

```python
stats.kstest(df.pop_growth_norm , 'norm')
#The p value is 0.78, we conclude that the population growth is normally distributed
```

### 3.4 Comparing normally distributed varibles for the clusters
#### 3.4.1 Analyzing the population growth of clusters using ANOVA

```python
##Testing the correlation between clusters and population growth
import statsmodels.api as sm
from statsmodels.formula.api import ols

#We hypothesize that there is a significant difference in the pop_growth of different clusters
#Since clusters are number, we want to categorize them during the test, hence the C()
mod = ols('pop_growth ~ C(cluster)',
                data=X).fit()
#Calculating aov table                
aov_table = sm.stats.anova_lm(mod)
print(aov_table)

#Calculating eta
esq_sm = aov_table['sum_sq'][0]/(aov_table['sum_sq'][0]+aov_table['sum_sq'][1])

print(esq_sm)
#eta is 0.41, we can conclude that there is some correlation between cluster and pop_growth
```

```python
#Verifying that another method returns the same results
def eta_squared(x,y):
    mean_y = y.mean()
    categories = []
    for category in x.unique():
        yi_category = y[x==category]
        categories.append({'ni': len(yi_category),
                        'mean_category': yi_category.mean()})
    TSS = sum([(yj-mean_y)**2 for yj in y])
    ESS = sum([c['ni']*(c['mean_category']-mean_y)**2 for c in categories])
    return ESS/TSS

eta_squared(X["cluster"], X["pop_growth"])
#We have the exact same result !
```

#### 3.4.2 Analyzing the Protein Supply of clusters using ANOVA

```python
##Checking if there's a correlation between protein supply and the cluster
import statsmodels.api as sm
from statsmodels.formula.api import ols

#We hypothesize that there is a significant difference in the ps_tot of different clusters
#Since clusters are number, we want to categorize them during the test, hence the C()
mod = ols('ps_tot ~ C(cluster)',
                data=X).fit()
#Calculating aov table                
aov_table = sm.stats.anova_lm(mod)
print(aov_table)

#Calculating eta
esq_sm = aov_table['sum_sq'][0]/(aov_table['sum_sq'][0]+aov_table['sum_sq'][1])

print(esq_sm)
#eta is 0.30, we can conclude that there is some correlation between cluster and gdp_growth
```

#### 3.4.3 Iterating over the clusters to find the highest correlation

```python
#Now we can try to find clusters with the highest amount of pop_growth correlation
arr = []
#Iterating over all clusters
for i in range(0,5,1):
    for j in range(1,5,1):
        #Avoid calculating eta for 2 identical clusters
        if i == j:
            continue
        #Check if i*j is in the array (to avoir duplicate calculations)
        elif np.in1d([i*j],arr).any():
            continue
        else:
            #If i!=0, we append the multiplication of both numbers to our array (to avoid duplication)
            #Since 0 is excluded from the range of j, we cannot have duplicates for i = 0
            if i != 0:
                arr.append([i*j])
            #Creating a subset with only the 2 selected clusters
            Xij = X[(X.cluster == i) | (X.cluster ==j)]
            print("--------------------------------------------------------")
            #Calculating Eta
            print("Eta of cluster {} and cluster {} for pop_growth is : {:.2f}".format(i,j,eta_squared(Xij["cluster"],Xij["pop_growth"])))
```
